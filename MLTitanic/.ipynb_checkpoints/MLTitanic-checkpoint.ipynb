```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import os
for dirname, _, filenames in os.walk('kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train = pd.read_csv('kaggle/input/titanic/train.csv')
```

```python
def impute_embr(emb):
    
    if pd.isnull(emb) or pd.isna(emb):
        emb = 77
    else:
        emb = ord(emb)
    emb = int(emb)
    
    return emb
```

```python
def prep_data(data):
    
    sex = pd.get_dummies(data['Sex'], drop_first=True)
    data['Sex'] = sex
    
    data['Embarked'] = data['Embarked'].apply(impute_embr)

    return data
```

```python
train = prep_data(train)
train
```

# Training and predicting data

```python
import numpy as np

from sklearn.compose import ColumnTransformer
from sklearn.datasets import fetch_openml
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, MaxAbsScaler, Normalizer, KernelCenterer, PowerTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.gaussian_process import GaussianProcessClassifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB      
from sklearn.decomposition import PCA, TruncatedSVD

numeric_features = ["Age", "Fare"]
numeric_transformer = Pipeline(
    steps=[("imputer", KNNImputer(n_neighbors=8, weights='distance')),
           ("scaler", MaxAbsScaler())]
)

categorical_features = ["Embarked", "Sex", "Pclass"]
categorical_transformer = Pipeline(
    steps=[("imputer", SimpleImputer(strategy='median')),
           ("norma", Normalizer()),
           ("encoder", OneHotEncoder(handle_unknown='ignore'))]
)


preprocessor = ColumnTransformer(
    transformers=[
        ("numb", numeric_transformer, numeric_features),
        ("catg", categorical_transformer, categorical_features),
    ]
)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import RadiusNeighborsClassifier 
from sklearn import svm

clf = Pipeline(
    steps=[("preprocessor", preprocessor), ("classifier", svm.LinearSVC(max_iter=10500))]
)

X_train, X_test, y_train, y_test = train_test_split(train.drop('Survived', axis=1), train['Survived'], test_size=0.2, random_state=6)

clf.fit(X_train, y_train)
print("model score: %.3f" % clf.score(X_test, y_test))
clf
```
