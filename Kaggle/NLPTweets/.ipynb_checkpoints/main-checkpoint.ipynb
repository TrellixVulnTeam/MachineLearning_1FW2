# 3. Data Transform

```python
# Delete all symbols that have low cost for ML

import re, string
def remove_url(text):
    url = re.compile(r'https?://\S+|www\.\S+')
    return url.sub(r'',text)

def remove_html(text):
    html=re.compile(r'<.*?>')
    return html.sub(r'',text)

def remove_new_line(text):
    new_line=re.compile(r'\n')
    return new_line.sub(r'',text)

def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)


def remove_atsymbol(text):
    name = re.compile(r'@\S+')
    return name.sub(r' ', text)


def remove_hashtag(text):
    hashtag = re.compile(r'#')
    return hashtag.sub(r' ', text)


def remove_exclamation(text):
    exclamation = re.compile(r'!')
    return exclamation.sub(r' ', text)


def remove_punc(text):
    return text.translate(str.maketrans('','',string.punctuation))
```

```python
# Shuffle Data
train_df = train_df.sample(frac=1, random_state=42)

# Tranform Text
train_df_cp = train_df.copy()
transform_methods = [remove_new_line, remove_url, remove_emoji, remove_atsymbol, remove_hashtag, remove_exclamation, lambda x: x.lower()]
for transform_method in transform_methods:
    train_df['text'] = train_df['text'].apply(transform_method)
```

```python
# visualize first transformation
k = list(zip(train_df["text"].values,train_df_cp["text"].values))[9:10]
print('String before:', k[0][1])
print('String after:', k[0][0])
```

# 4. Train Test split

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(test_size=0.2 ,random_state=42)

for train_index, test_index in split.split(train_df, train_df["target"]):
  X_train = train_df["text"][train_index]
  y_train = train_df["target"][train_index]
  X_val = train_df["text"][test_index]
  y_val= train_df["target"][test_index]
```

# 5. Training

```python
encoder_url = 'https://tfhub.dev/google/universal-sentence-encoder/4'

import tensorflow_hub as hub
embed = hub.load(encoder_url)
embed_samples = embed(['There is a flood in my street'])
```

```python
import tensorflow as tf

sentence_layer_encoder = hub.KerasLayer(encoder_url,
                                        input_shape=[], dtype=tf.string,
                                        trainable=False, name="USE")
model = tf.keras.Sequential(
    [
     sentence_layer_encoder,
     tf.keras.layers.Dense(64, activation="elu"),
     tf.keras.layers.Dense(128, activation="gelu"),
     tf.keras.layers.Dense(1, activation="sigmoid")
    ], name="ELU_GELU"
)
model.summary()
```

```python
model.compile(loss="binary_crossentropy",  
                optimizer="rmsprop", metrics=["accuracy"])
model_history = model.fit(x=X_train,
                              y=y_train,
                              epochs=5,
                              validation_data=[X_val,y_val])
model.evaluate(X_val,y_val)
model_predict_proba = model.predict(X_val)
```

```python
from pandas.compat.numpy.function import process_skipna
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def calculate_results(y_true, y_pred):

  accuracy = accuracy_score(y_true, y_pred)
  precision, recall, f1score , _ = precision_recall_fscore_support(y_true, y_pred, average="weighted")
  model_results = { "accuracy":[accuracy],
                    "precision":[precision],
                    "recall":[recall],
                    "f1score":[f1score],
                   }
  return model_results

from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(y_pred, y_test):
    
    pred_classes = ['NotDisaster','Disaster']
    
    cm = confusion_matrix(y_pred,y_test)

    
    # plot confusion matrix
    plt.figure(figsize=(5,3))
    sns.heatmap(cm, cmap=plt.cm.Blues, cbar=False, annot=True, fmt='d', 
                xticklabels=pred_classes,
                yticklabels=pred_classes)
    plt.title(f'Confusion matrix', fontsize=16)
    
```

```python
model_pred = tf.squeeze(tf.round(model_predict_proba))
model_results = calculate_results(y_val, model_pred)

score_df = pd.DataFrame((pd.DataFrame.from_dict(model_results)))
score_df
```

```python
plot_confusion_matrix(model_pred, y_val)
```

```python
val_df = pd.DataFrame({"text": X_val,
                       "target": y_val,
                       "pred": model_pred,
                       "pred_prob": tf.squeeze(model_predict_proba)})
val_df.head(12)
```

```python
most_wrong = val_df[val_df["target"] != val_df["pred"]].sort_values("pred_prob", ascending=False)
print('Total wrong predicions ',len(most_wrong), ' out of ', len(val_df))
sns.histplot(data=most_wrong, x="pred_prob", hue="target");
```

## 6. Prediction

```python
# Making predictions on the test dataset
for transform_method in transform_methods:
    test_df['text'] = test_df['text'].apply(transform_method)
```

```python
preds = tf.round(tf.squeeze(model.predict(test_df['text']), axis=1)) # average and round the prediction probabilities to get prediction classes
test_df['target'] = preds
test_df['target'] = test_df['target'].astype('int')
```
